# Claude with Google Vertex AI - API Requests

Making basic API requests to Claude models via Google Cloud Vertex AI using the Anthropic Python SDK.

```python
# Create an API client
import os
from dotenv import load_dotenv
from anthropic import AnthropicVertex

load_dotenv()

client = AnthropicVertex(
    region=os.environ["GCP_REGION"],
    project_id=os.environ["GCP_PROJECT_ID"],
)
model = "claude-sonnet-4@20250514"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None):
    parameters = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    if system:
        parameters["system"] = system
    
    message = client.messages.create(**parameters)
    
    return message.content[0].text
```

```python
messages = []

system = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """

add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")

answer = chat(messages, system=system)

print(answer)
```
