````python
import sys
sys.path.insert(0, "..")

from utils import setup_client, add_user_message, add_assistant_message, chat

client, model = setup_client()
````

````python
messages = []

add_user_message(
    messages,
    "Is tea or coffee better at breakfast?",
)

add_assistant_message(
    messages,
    "Coffee is better because"
)

answer = chat(client, model, messages)

print(answer)
````

````python
messages = []

add_user_message(
    messages,
    "Count from 1 to 10",
)

answer = chat(client, model, messages, stop_sequences=[", 5"])

print(answer)
````

````python
messages = []

add_user_message(
    messages,
    "Generate a very short event bridge rule as json"
)
add_assistant_message(
    messages,
    "```json"
)

text = chat(client, model, messages, stop_sequences=["```"])
````

````python
import json

parsed_json = json.loads(text.strip())

print(parsed_json)
````
